```python
import pandas as pd
import sys
import os

# Add parent directory to path for local development
sys.path.insert(0, os.path.dirname(os.path.dirname(os.path.abspath(__file__))))

from easygoogle import Sheets

# Initialize Sheets manager
# Credentials will be auto-detected from:
# 1. GS_UTILS_JSON_FOLDER environment variable
# 2. .easygoogle_config.yaml file
# 3. Default: .secret/ folder
sheets = Sheets()

# Or specify custom path:
# sheets = Sheets(json_folder="../.secret")
```

```python
# 구글 시트 URL 및 시트 이름을 지정하세요.
spreadsheet_id = "spreadsheet_id"
spreadsheet_url = f"https://docs.google.com/spreadsheets/d/{spreadsheet_id}"
```

```python
# 1. 시트 이름과 ID 딕셔너리 얻기
name_id_dict = sheets.get_sheet_name_id_dict(spreadsheet_url)
print(name_id_dict)
```

```python
# 2. 시트 이름 리스트 얻기
sheet_names = sheets.get_sheet_name_list(spreadsheet_url)
print(sheet_names)
```

```python
# 3. 시트에서 데이터프레임 불러오기
sheet_name = "Sheet1"
df = sheets.get_dataframe_from_sheet(spreadsheet_url, sheet_name)
print(df)
```

```python
# 4. 데이터프레임을 시트에 쓰기 (기존 데이터 삭제 후 입력)
new_df = pd.DataFrame({
    "이름": ["홍길동", "김철수"],
    "나이": [30, 25]
})
sheets.clear_and_set_worksheet(spreadsheet_url, "새시트", new_df)
```
